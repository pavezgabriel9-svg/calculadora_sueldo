```python
import bcchapi
import pandas as pd 
import os 
from datetime import date
from dotenv import load_dotenv
import numpy as np 
load_dotenv(override=True)
```

```python
hoy = date.today()
email = os.getenv("email")
password = os.getenv("password")
siete = bcchapi.Siete(email, password)
```

```python
df = siete.cuadro(
  series=["F073.UFF.PRE.Z.D"],
  nombres = ["UF diaria"],
  desde=hoy,
  hasta=hoy,
  variacion=0,
  frecuencia="D",
  observado={"UF diaria":'last'}
)

df = df.reset_index()
df.rename(columns={'index':'fecha'}, inplace=True)
df
```
